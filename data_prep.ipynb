### tvtropes.clusters.txt

```python
tvt = pd.read_csv('./MovieSummaries/tvtropes.clusters.txt', sep='\t', header=None)
tvt.head()
```

```python
tvt_cols = [
    'character_types',
    'instances'
]
tvt.columns = tvt_cols
```

```python
# test
tvt.head()
```

```python
tvt['instances'][0]
```

```python
# convert str to dict
tvt['instances'] = tvt['instances'].apply(json.loads)
```

```python
# test
tvt['instances'][0]
```

```python
# extract json
```

```python
tvt_df = pd.json_normalize(tvt['instances'], max_level=1)
tvt_df.head()
```

```python
tvt_df['character_types'] = tvt['character_types']
```

```python
# text
tvt_df.head()
```

```python
# rename columns
tvt_cols = ['char', 'movie', 'Freebase movie ID', 'actor', 'character_types']

tvt_df.columns = tvt_cols
```

```python
# test
tvt_df.head()
```
